# Exploratory Data Analysis

# Importing the Titanic Dataset

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
df.info()
```

```python
missing_values = df.isna().sum()
missing_values
```

## Percentage of missing values

```python
percentage_missing = 100* missing_values / len(df)
percentage_missing
```

#### Now from above output it is evident that 77% of the `Cabin` data is missing which is huge. On the other hand, almost 19% of `Age` and 0.2% of `Embarked` data is also missing.

Below we have the visulaization for the same.

```python
import missingno as msno
```

```python
msno.matrix(df);
```

## Handling Missing Values

```python
df.dropna(subset=['Embarked'], how='any',inplace=True)
msno.matrix(df);
```

### Filling the `Age` Null values by the mean 

```python
df['Age'] = df['Age'].replace(np.NAN,df['Age'].mean())
df['Age'].isnull().sum()
```

```python
msno.matrix(df);
```

### Handling missing values for `Cabin` variable i.e. categorical by filling the `NAN` places with `Unknown` text

```python
df['Cabin'] = df['Cabin'].fillna('Unknown')
df['Cabin'].isnull().sum()
```

### Lets plot the graph again as we expect no missing values across our dataset

```python
msno.matrix(df);
```

```python
df.describe()
```

# Hypothesis using Chi-Square test

## Pclass vs Survived

```python
from scipy.stats import chi2_contingency
from scipy.stats import chi2
```

```python
# Getting contingency table
contingency_table = pd.crosstab(df['Pclass'], df['Survived'])
print(contingency_table)

# Chi-square test of independence.
stat, p, dof, expected = chi2_contingency(contingency_table)


# interpret test-statistic
print('P-Value is: ', p)
alpha = 0.05
if p <= alpha:
    print('Null Hypothesis is Rejected, meaning both variables are dependent')
else:
    print('Failed to reject the Null Hypothesis, meaning both varibales are independent')
```

```python
df[['Pclass','Survived']].groupby(['Pclass'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```

```python
sns.catplot(x='Survived',col='Pclass',kind='count',data=df);
```
